```python
import pandas as pd

# Reading the CSV files
student_data = pd.read_csv('student_data.csv')
classroom_analytics = pd.read_csv('classroom_analytics.csv')
learning_analytics = pd.read_csv('learning_analytics.csv')
motivation_metrics = pd.read_csv('motivation_metrics.csv')
performance_metrics = pd.read_csv('performance_metrics.csv')
planets_analytics = pd.read_csv('planets_analytics.csv')
user_experience_metrics = pd.read_csv('user_experience_metrics.csv')

# Display the first few rows of each DataFrame to verify successful loading
print(student_data.head())
print(classroom_analytics.head())
print(learning_analytics.head())
print(motivation_metrics.head())
print(performance_metrics.head())
print(planets_analytics.head())
print(user_experience_metrics.head())
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define table relationships based on the common columns
relationships = {
    "student_data": ["learning_analytics", "motivation_metrics", "performance_metrics", "user_experience_metrics"],
    "classroom_analytics": ["planets_analytics"],
    "planets_analytics": ["classroom_analytics"],  # Connected through classroom_id
}

# Create a directed graph
G = nx.DiGraph()

# Add nodes (tables)
for table in relationships.keys():
    G.add_node(table, color="lightblue")  # Main tables
for sub_table_list in relationships.values():
    for sub_table in sub_table_list:
        G.add_node(sub_table, color="lightgreen")  # Related tables

# Add edges (relationships)
for main_table, sub_table_list in relationships.items():
    for sub_table in sub_table_list:
        G.add_edge(main_table, sub_table)

# Create a layout for the graph
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42, k=1.5)  # Adjust 'k' for better spacing

# Get node colors
colors = [G.nodes[node]['color'] for node in G.nodes]

# Draw the graph
nx.draw(G, pos, with_labels=True, node_color=colors, node_size=3000, edge_color="gray", font_size=10, font_weight="bold", alpha=0.7)

# Title and show the plot
plt.title("Entity Relationship Diagram (ERD) - Datasets Relationships")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the performance_metrics dataset
performance_metrics = pd.read_csv("performance_metrics.csv")

# Compute the correlation matrix
correlation_matrix = performance_metrics.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Performance Metrics")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the figure size for multiple plots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Scatter Plot: Lesson Completion Rates vs. Reading Comprehension Scores
sns.regplot(x=performance_metrics["lesson_completion_rates"], 
            y=performance_metrics["reading_comp_scores"], 
            ax=axes[0, 0], scatter_kws={"alpha": 0.5})
axes[0, 0].set_title("Lesson Completion Rates vs. Reading Comprehension Scores")

# Scatter Plot: Game Scores vs. Reading Comprehension Scores
sns.regplot(x=performance_metrics["game_scores_for_mini_games"], 
            y=performance_metrics["reading_comp_scores"], 
            ax=axes[0, 1], scatter_kws={"alpha": 0.5})
axes[0, 1].set_title("Game Scores vs. Reading Comprehension Scores")

# Scatter Plot: Frequency of Reading Strategy Usage vs. Reading Comprehension Scores
sns.regplot(x=performance_metrics["frequency_of_reading_strategy_usage"], 
            y=performance_metrics["reading_comp_scores"], 
            ax=axes[1, 0], scatter_kws={"alpha": 0.5})
axes[1, 0].set_title("Reading Strategy Usage vs. Reading Comprehension Scores")

# Scatter Plot: Frequency of Reading Strategy Usage vs. Lesson Completion Rates
sns.regplot(x=performance_metrics["frequency_of_reading_strategy_usage"], 
            y=performance_metrics["lesson_completion_rates"], 
            ax=axes[1, 1], scatter_kws={"alpha": 0.5})
axes[1, 1].set_title("Reading Strategy Usage vs. Lesson Completion Rates")

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the performance_metrics dataset
performance_metrics = pd.read_csv("performance_metrics.csv")

# Compute the correlation matrix for only the two key columns
correlation_matrix = performance_metrics.corr()[["game_scores_for_mini_games", "reading_comp_scores"]]

# Plot the heatmap
plt.figure(figsize=(6, 6))
sns.heatmap(correlation_matrix.T, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation of Game Scores & Reading Comprehension Scores with Other Features")
plt.xlabel("Metrics")
plt.ylabel("Features")
plt.show()
```
